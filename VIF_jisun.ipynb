```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 수치형 변수를 선택
numeric_columns = aug_train.select_dtypes(include=['int64', 'float64']).columns
```

```python
# VIF 데이터프레임 생성
vif_data = pd.DataFrame()
vif_data["feature"] = numeric_columns
vif_data["VIF"] = [variance_inflation_factor(aug_train[numeric_columns].values, i) for i in range(len(numeric_columns))]

# VIF 값을 기준으로 내림차순 정렬
vif_data_sorted = vif_data.sort_values(by='VIF', ascending=False)

# 결과 출력
print(vif_data_sorted)
```
